### R-CNN (Regions with Convolutional Neuron Networks features)
image.png

<출처 : Rich feature hierarchies for accurate object detection and semantic segmentation
Tech report (v5) >

- 데이터와 bounding box로 된 label 입력
- input image에서 후보 sub-image 추출
- 각 이미지 CNN을 통한 특징 추출
- SVM을 통한 분류
- CNN을 통과해서 나온 Bounding Box, 처음에 지정한 label 간의 차이를 줄이기 위한 Regression 진행
- 최초 라벨에 일치하도록 조정

**단점**

- 복잡한 모델
- 너무 오래 걸리는 시간


### Fast R-CNN
- ROI 영역을 만든다
- 전체 이미지를 한 번 CNN에 통과시킨다
- CNN을 통과된 이미지에서 기존에 만든 ROI를 찾는다

image.png

<출처 - https://yeomko.tistory.com/15>

**개선점**
- 2000번의 CNN 연산에서 한 번의 CNN 연산으로 줄여 속도가 빨라짐

**단점**
- 하지만 여전히 ROI를 찾는 selective search가 CNN 외부에서 이루어지므로 실시간 처리를 하기에는 느림

### Faster R-CNN
- Region proposal Networks(RPN)을 CNN과 ROI 추출층 사이에 넣어서 Selective Search 또한 CNN 연산 안에서 이루어지도록 바꿈

### YOLO (You Only Look Once)
image.png

- 이미지를 n*n개의 그리드로 나눈다
- 이미지를 CNN에 통과시키면, confidence score와 bounding box를 얻는다
- bounding box와 confidence score가 최대한으로 겹치는 지점을 찾는다
