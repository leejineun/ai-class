### Numpy
- Python에서 벡터, 행렬 등 수치 연산을 수행하는 선형대수 라이브러리
- C로 구현되어 속도가 빠름

```python
import numpy as np
```

- shape : 배열의 각 축의 크기
- ndim : 축의 개수
- dtype : 각 요소의 타입
    - int
    - int16
    - int32
    - int64
    - float
    - float16
    - float32
    - float64
    - bool
    등
- size : 전체 요소의 개수

```python
a = np.zeros((3, 5))
print(a)
```

```python
print(type(a))
```

```python
print('shape', a.shape)
print('ndim', a.ndim)
print('dtype', a.dtype)
print('size', a.size)
```

```python
b = np.array([1, 2, 3])
print(b.ndim)
```

### Methods

- np.zeros
 - 0으로 가득찬 array를 만든다


```python
a = np.zeros(3)
print(a, '\n', a.shape)
```

```python
a = np.zeros(3, dtype='int64')
print(a, '\n', type(a[0]))
```

```python
a = np.zeros((3, 5))
print(a, '\n', a.shape)
```

```python
# (4, 2, 2) shape을 가진 다차원 배열 생성
# dtype=int32
a = np.zeros((4,2,2), dtype='int32')
print(a, '\n', type(a[0][0][0]))
```

- np.ones
    - 1로 가득찬 array를 만든다

```python
a = np.ones(3)
print(a, '\n', a.shape)
```

```python
a = np.ones(3)
print(a.dtype)
a = np.ones(3, dtype='int32')
print(a.dtype)
```

```python
a = np.ones((3, 5))
print(a, '\n', a.shape)
```

```python
list_1 = [1, 2, 3] * 3
print(list_1)
```

```python
array_1 = np.array([1,2,3]) * 3
print(array_1)
```

```python
a = np.ones((3, 5)) * 5
print(a)
```

- np.empty()

```python
np.empty((3,3))
```

- np.arange
    - 특정 범위의 array를 만든다
    - 간격을 따로 지정하지 않으면, 모든 간격은 1이다

```python
a = np.arange(10)
print(a)
```

```python
a = np.arange(0, 30, 3)
print(a)
```

```python
a = np.arange(30, 50)
print(a)
```

```python
a = np.arange(10, 20, 2) * 3
print(a)
```

- np.full
    - 지정한 수로 가득한 array를 만든다

```python
a = np.full((2, 2), 7)
print(a)
```

- 단위행렬
    - 모든 대각선 원소가 1, 나머지는 0인 행렬 생성

```python
a = np.eye(4)
print(a, '\n', a.shape)
```

- 삼각행렬

```python
a = np.tri(10)
print(a, '\n', a.shape)
```

- _like
    - 지정된 배열과 shape이 같은 행렬 생성

```python
a = np.array([[1, 2, 3], [4, 5, 6]])
print(a, '\n', a.shape)
```

```python
a = np.zeros_like(a)
print(a, '\n', a.shape)
```

```python
a = np.ones_like(a)
print(a, '\n', a.shape)
```

```python
a = np.ones_like(a) * 6
print(a, '\n', a.shape)
```

```python
a = np.empty_like(a)
print(a, '\n', a.shape)
```

- np.random.rand
    - 0에서 1사이의 랜덤 값을 가진 array를 반환한다

```python
a = np.random.rand(3,2)
print(a, '\n', a.shape)
```

- np.random.randn
    - 0에서 1사이의 표준정규분포 값을 랜덤으로 얻는다.

```python
a = np.random.randn(3, 3)
print(a, '\n', a.shape)
```

- np.random.randint
    - 정수 랜덤 값만을 가진 array를 반환한다
    - 하나 이상의 하이퍼 파라미터를 반드시 적어야 한다
    - low만 적을 경우, low를 high로 인식한다

```python
np.random.randint(5)
```

```python
np.random.randint(2, 5)
```

```python
np.random.randint(2, 5, size=(3,3))
```

- np.random.random
    - rand와 거의 동일하다

```python
np.random.random((2, 3))
```

- np.random.uniform
    - 특정 범위의 랜덤 값을 가진 array를 만든다


```python
print(np.random.uniform(0 ,100, size=(2, 2)))
```
