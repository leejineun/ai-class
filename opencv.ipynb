```python
# canny edge
cap = cv2.VideoCapture('./videos/test_video.webm')

while (cap.isOpened()) : 
    ret, frame = cap.read()
    
    dst = frame.copy()
    dst = cv2.cvtColor(dst, cv2.COLOR_BGR2GRAY)

    canny = cv2.Canny(dst, 10, 255)
    
    cv2.imshow('Video', canny)
    
    if cv2.waitKey(1) & 0xFF == ord('q') : 
        break
        
cap.release()
cv2.destroyAllWindows()
```

```python
def onChange(x) : 
    pass

def get_roi(frame, x, y, w, h) :
    roi = frame[y:y+h, x:x+w]
    return roi


cap = cv2.VideoCapture('./videos/test_video.webm')
index = 0
time_marker = time.time()

while (cap.isOpened()) : 
    ret, frame = cap.read()
    
    roi = get_roi(frame, 300, 0, 750, 200)
    
    laplacian = cv2.Laplacian(roi, -1)
    dst = laplacian.copy()
    
    gray = cv2.cvtColor(dst, cv2.COLOR_BGR2GRAY)
    
    
    circles = cv2.HoughCircles(gray, cv2.HOUGH_GRADIENT, 1, 200, param1=300, param2=18, minRadius=15, maxRadius=20)
    
    timestamp = time.time()
    try :
        for i in circles[0] : 
            cv2.circle(gray, (i[0], i[1]), i[2], (0, 255, 255), 3)
            print('Circle Detect')
            if timestamp - time_marker > 2 : 
                print('Save')
                cv2.imwrite('./images/lol_{}.jpg'.format(index), frame)
                index += 1
                time_marker = time.time()
                
    except Exception : 
        pass
    
    cv2.imshow('Video', roi)
    
    if cv2.waitKey(2) & 0xFF == ord('q') : 
        break
        
cap.release()
cv2.destroyAllWindows()
```
