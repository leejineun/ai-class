# LeNet
- CNN계의 고전
- 우편번호와 수표의 필기체를 인식하기 위해 만든 단순한 구조의 모델


# AlexNet

- 2012 ImageNet 우승 모델
- 이미지 인식 방식을 DL로 전환
- 5개의 convolution layer, 3개의 FC layer

# InceptionV3
- 가장 많이 사용하는 모델 중 하나
- 비대칭 합성곱을 사용하여 연산량 줄임
- 보조 분류기 사용



# VGGNET

- 2014년 ImageNet 준우승
- 옥스포드 대학의 연구팀에서 개발
- 16개 또는 19개 층으로 구성된 모델을 주로 사용 (VGG16, VGG19)
- VGGNET을 기점으로 레이어가 급격히 증가

### 작은 필터로 여러 번 컨볼루션 하는 것이 큰 필터로 한 번 하는 것보다 나은 점
- 작은 필터로 여러번 할 경우 가중치가 더 적다
- 가중치가 많으면 훈련시켜야 할 가중치가 많다는 의미
- 즉, 학습 속도가 느려진다.
- 또한 필터가 작아서 층이 늘어나면 특성에 비선형성을 증가시켜 데이터 특성이 점점 더 유용해진다
