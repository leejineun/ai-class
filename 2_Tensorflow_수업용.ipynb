# Tensorflow
### Tensorflow란?
- 구글에서 만든 오픈소스 라이브러리
- 텐서 연산을 지원
- 2.x 버전으로 오면서 문법이 쉬워짐

### 텐서란?
- dtype을 가진 다차원 배열 
- 일종의 ``np.arrays``와 같다

### 용어
- type : ``int8``, ``int16``, ``float32``, ``string`` 등
- 형상(shape): 텐서의 각 차원의 길이(요소의 수)  (3, 2, 4, 5)
- 순위(rank): 텐서 차원의 수입니다. 스칼라는 순위가 0이고 벡터의 순위는 1이며 행렬의 순위는 2입니다.
- 축 또는 차원(axis): 텐서의 특정 차원
- 크기(size): 텐서의 총 항목 수, 곱 형상 벡터
